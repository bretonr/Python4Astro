#### Regular numerical integration

Numerical integrations (using quadrature) are possible on user-defined functions.

```python
## Defining a function
def gaussian(x, A=1, x0=0, sigma=1):
    return A*np.exp(-0.5*(x-x0)**2/sigma**2)/sigma/np.sqrt(2*np.pi)

val, abserr = scipy.integrate.quad(gaussian, -np.inf, np.inf)
print("integral value = {}, absolute error = {}".format(val, abserr))

## We can even pass the optional arguments to the function
val, abserr = scipy.integrate.quad(gaussian, -np.inf, np.inf, args=(1.5, 0, 2))
print("integral value = {}, absolute error = {}".format(val, abserr))
```

#### Solving ordinary differential equations (ODEs)

Let's consider the double compound pendulum (see the full info on the [double pendulum](http://en.wikipedia.org/wiki/Double_pendulum) Wikipedia page).

```python
Image(url='http://upload.wikimedia.org/wikipedia/commons/c/c9/Double-compound-pendulum-dimensioned.svg')
```

The equations of motion of the pendulum are given on the wiki page:

${\dot \theta_1} = \frac{6}{m\ell^2} \frac{ 2 p_{\theta_1} - 3 \cos(\theta_1-\theta_2) p_{\theta_2}}{16 - 9 \cos^2(\theta_1-\theta_2)}$

${\dot \theta_2} = \frac{6}{m\ell^2} \frac{ 8 p_{\theta_2} - 3 \cos(\theta_1-\theta_2) p_{\theta_1}}{16 - 9 \cos^2(\theta_1-\theta_2)}.$

${\dot p_{\theta_1}} = -\frac{1}{2} m \ell^2 \left [ {\dot \theta_1} {\dot \theta_2} \sin (\theta_1-\theta_2) + 3 \frac{g}{\ell} \sin \theta_1 \right ]$

${\dot p_{\theta_2}} = -\frac{1}{2} m \ell^2 \left [ -{\dot \theta_1} {\dot \theta_2} \sin (\theta_1-\theta_2) +  \frac{g}{\ell} \sin \theta_2 \right]$

To make the Python code simpler to follow, let's introduce new variable names and the vector notation: $x = [\theta_1, \theta_2, p_{\theta_1}, p_{\theta_2}]$

${\dot x_1} = \frac{6}{m\ell^2} \frac{ 2 x_3 - 3 \cos(x_1-x_2) x_4}{16 - 9 \cos^2(x_1-x_2)}$

${\dot x_2} = \frac{6}{m\ell^2} \frac{ 8 x_4 - 3 \cos(x_1-x_2) x_3}{16 - 9 \cos^2(x_1-x_2)}$

${\dot x_3} = -\frac{1}{2} m \ell^2 \left [ {\dot x_1} {\dot x_2} \sin (x_1-x_2) + 3 \frac{g}{\ell} \sin x_1 \right ]$

${\dot x_4} = -\frac{1}{2} m \ell^2 \left [ -{\dot x_1} {\dot x_2} \sin (x_1-x_2) +  \frac{g}{\ell} \sin x_2 \right]$

```python
g = 9.82
L = 0.5
m = 0.1

def dx(x, t):
    """
    The right-hand side of the pendulum ODE
    """
    x1, x2, x3, x4 = x[0], x[1], x[2], x[3]
    
    dx1 = 6.0/(m*L**2) * (2 * x3 - 3 * cos(x1-x2) * x4)/(16 - 9 * cos(x1-x2)**2)
    dx2 = 6.0/(m*L**2) * (8 * x4 - 3 * cos(x1-x2) * x3)/(16 - 9 * cos(x1-x2)**2)
    dx3 = -0.5 * m * L**2 * ( dx1 * dx2 * sin(x1-x2) + 3 * (g/L) * sin(x1))
    dx4 = -0.5 * m * L**2 * (-dx1 * dx2 * sin(x1-x2) + (g/L) * sin(x2))
    
    return [dx1, dx2, dx3, dx4]
```

```python
## choose an initial state
x0 = [pi/4, pi/2, 0, 0]

## time coodinate to solve the ODE for: from 0 to 10 seconds
t = linspace(0, 10, 250)

## solve the ODE problem
x = scipy.integrate.odeint(dx, x0, t)

## what does the result look like?
print("the solution to the ODE has a shape = {}".format(x.shape))
```

#### Spline interpolation

Let's interpolate the previous results, $\theta_1$ and p$_1$, as a spline function which can then be evalutated anywhere.

```python
## calculate the interpolation coefficients
tck_theta1 = scipy.interpolate.splrep(t, x[:, 0])
tck_p1 = scipy.interpolate.splrep(t, x[:, 2])

## create a convenience function to return the values
theta1 = lambda t: scipy.interpolate.splev(t, tck_theta1)
p1 = lambda t: scipy.interpolate.splev(t, tck_p1)

## plot the results
## here we will stack the panels to save space
fig = pylab.figure(figsize=(12,8))

## add axes takes a list, [l,b,w,h], in units of fractional figure
ax0 = fig.add_axes([0.1,0.1,0.85,0.425])
ax1 = fig.add_axes([0.1,0.525,0.85,0.425])

new_t = np.linspace(0,10,400)

ax0.scatter(t, x[:, 0], marker='o', facecolor='none', edgecolor='b')
ax0.plot(new_t, theta1(new_t), 'r')
ax0.plot(t, t*0, 'k--')
ax0.set_xlabel('time')
ax0.set_ylabel('angle (radian)')
ax0.set_ylim([-2.1,2.1])
ax0.set_xlim([0,10])
ax1.scatter(t, x[:, 2], marker='o', facecolor='none', edgecolor='b')
ax1.plot(new_t, p1(new_t), 'r')
ax1.plot(t, t*0, 'k--')
ax1.set_xticklabels([])
ax1.set_yticks(np.linspace(-0.15,0.15,7))
ax1.set_xlim([0,10])
```

#### Finding the minimum of a function

Let's consider our previous problem. Let's find the time at which the pendulum is at minimum angle.

```python
theta1_min = scipy.optimize.fmin_bfgs(theta1, 0.7)
print("The first minimum of the position is theta1 = {:.4}".format(theta1_min[0]))
```

#### Finding the root of a function

Let's find the time at which the momentum vanishes.

```python
## fsolve takes an initial guess. One could also use Newton's method, aslo part of scipy.optimize.
p1_min = scipy.optimize.fsolve(p1, 0.7)
## In some cases, it is easier to solve using a bisection and providing bounds (this might be better when there are several minima)
p1_min2 = scipy.optimize.bisect(p1, 0.6, 0.9)

print("The first root of the momentum is p1 = {:.4} (fsolve), and p1 = {:.4} (bisect)".format(p1_min[0], p1_min2))
```

#### Curve fitting

In principle one could use *scipy.optimize.fmin* to fit a curve since the best-fit chi-square is a minimum. In practice, this works sometimes but it often fails especially if there are more than one parameters to fit, then likelihood is very smooth and the starting point near the best-fit solution. The other problem is that *fmin* doesn't provide any error on the fitted parameters.

Let's examine the case of a linear least-square fit first. We will attempt to fit a Gaussian to some mock data.

```python
## Redefining the Gaussian, without the normalisation factor
def gaussian(x, A=1, x0=0, sigma=1):
    return A*np.exp(-0.5*(x-x0)**2/sigma**2)

## First, we generate the mock data.
x = np.linspace(-10.,10.,101)
y_nonoise = gaussian(x, A=4.321, x0=0.0, sigma=1.0) + 1.234
y = y_nonoise + np.random.normal(scale=0.5, size=x.size)

## Our independent values are a Gaussian with an amplitude of 1
y_model = gaussian(x, A=1.0, x0=0.0, sigma=1.0)
```

We will use the routine *numpy.linalg.lstsq*, which solves the equation $ax = b$ by computing a vector $x$ that minimizes the Euclidean 2-norm $|| b - a x ||^2$

```python
def fit_linear(y, x=None, err=1.0, m=None, b=None):
    """
    Fit a line y = mx + b

    Input:
    y (array): data vector to be fitted
    x (array): independent data vector to use for the fit.
        If None, will just use the position index.
    err (float, array): errors on the data. Can be a float
        if they are the same for all, or an array.
    m (float): Value of the m parameter. If provided, will
        not be fitted.
    b (float): Value of the b parameter. If provided, will
        not be fitted.

    Output:
    (sol, res, rank, s)
    sol -> [b, m]
    """
    if x is None:
        x = np.arange(y.size, dtype=float)
    if (b is not None) and (m is not None):
        sol = [b, m]
        res = (((b + m*x - y)/err)**2).sum()
        rank = 0.
        s = 0.
    else:
        if b is not None:
            A = np.reshape(x/err,(x.shape[0],1))
            y1 = y-b
            y1 /= err
            sol, res, rank, s = scipy.linalg.lstsq(A, y1)
            sol = [b,sol[0]]
        elif m is not None:
            A = np.resize(1/err,(x.shape[0],1))
            y1 = y-m*x
            y1 /= err
            sol, res, rank, s = scipy.linalg.lstsq(A, y1)
            sol = [sol[0],m]
        else:
            A = (np.vstack([np.ones(x.size, dtype=float),x])/err).T
            y1 = y/err
            sol, res, rank, s = scipy.linalg.lstsq(A, y1)
    return sol, res, rank, s

(m,b), res, rank, s = fit_linear(y, y_model, m=None, b=None, err=0.5)
print("The best-fit parameters are m = {:.3f} and b = {:.3f}".format(m,b))
```

The problem with this routine is that it doesn't provide the best-fit parameter errors nor the chi2. Instead, we use our own version written from numerial recipe.

```python
def fit_linear_nr(y, x=None, err=1.0, m=None, b=None):
    """
    Fit a line y = mx + b

    Input:
    y (array): data vector to be fitted
    x (array): independent data vector to use for the fit.
        If None, will just use the position index.
    err (float, array): errors on the data. Can be a float
        if they are the same for all, or an array.
    m (float): Value of the m parameter. If provided, will
        not be fitted.
    b (float): Value of the b parameter. If provided, will
        not be fitted.

    Output:
    m, b, err_m, err_m, chi2
    """
    if isinstance(err, float):
        err = np.ones(y.size)*err
    if x is None:
        x = np.arange(y.size)
    err2 = err**2
    if (m is None) and (b is None):
        S = np.sum(1/err2,0)
        Sx = np.sum(x/err2,0)
        Sy = np.sum(y/err2,0)
        t = (x-Sx/S)/err
        Stt = np.sum(t**2,0)
        m = np.sum(t*y/err,0)/Stt
        b = (Sy-Sx*m)/S
        err_b = np.sqrt((1+Sx**2/(S*Stt))/S)
        err_m = np.sqrt(1/Stt)
    elif (m is not None) and (b is None):
        S = numpy.sum(1/err2,0)
        Sx = numpy.sum(x/err2,0)
        Sy = numpy.sum(y/err2,0)
        b = (Sy-m*Sx)/S
        err_b = numpy.sqrt(1/S)
        err_m = 0.0
    elif (m is None) and (b is not None):
        Sx = numpy.sum(x/err2,0)
        Sxx = numpy.sum(x**2/err2,0)
        Sxy = numpy.sum(x*y/err2,0)
        m = (Sxy-b*Sx)/Sxx
        err_m = numpy.sqrt(1/Sxx)
        err_b = 0.0
    else:
        err_m = 0.0
        err_b = 0.0
    chi2 = np.sum(((m*x+b - y)/err)**2,0)
    return m, b, err_m, err_b, chi2

m, b, err_m, err_b, chi2 = fit_linear_nr(y, y_model, m=None, b=None, err=0.5)
print("The best-fit parameters are m = {:.3f} +/- {:.3f} and b = {:.3f} +/- {:.3f}".format(m,err_m,b,err_b))
print("The best-fit chi2 is {:.3}".format(chi2))
```

#### Curve fitting of a non-linear model

To fit a non-linear model that is not "too messy", one can use the Levenberg-Marquardt algorigthm implemented in *scipy.optimize.curve_fit*. The function returns a list of the best-fit parameters and the covariance matrix. The error on the parameters can be inferred from the square root of the diagonal.

```python
## First, we generate the mock data.
x = np.linspace(-10.,10.,101)
y_nonoise = gaussian(x, A=4.321, x0=0.0, sigma=1.0)
y = y_nonoise + np.random.normal(scale=0.5, size=x.shape)
err_y = np.ones_like(x)*0.5

sol, cov = scipy.optimize.curve_fit(gaussian, x, y, p0=[4.1, 0., 1.], sigma=err_y)
print("The best-fit parameters are: A = {0[0]:.3f} +/- {1[0]:.3f}, x0 = {0[1]:.3f} +/- {1[1]:.3f}, sigma = {0[2]:.3f} +/- {1[2]:.3f}".format(sol, np.sqrt(cov.diagonal())))
chi2 = (((y-gaussian(x, *sol))/err_y)**2).sum()
print("The best-fit chi2 is {:.3f}".format(chi2))
```

```python
## plot the results
## here we will stack the panels to save space
fig, ax = pylab.subplots()

ax.scatter(x, y, marker='o', facecolor='none', edgecolor='b')
ax.plot(x, y_nonoise, 'r')
ax.plot(x, gaussian(x, *sol), 'g')
```
