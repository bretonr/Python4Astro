### Array broadcasting

Let's look at how basic operations (+, -, *, /) work on numpy arrays. This is called the array broadcasting rules. This section is taken from http://docs.scipy.org/doc/numpy/user/basics.broadcasting.html.

#### Broadcasting

The term broadcasting describes how numpy treats arrays with different shapes during arithmetic operations. Subject to certain constraints, the smaller array is “broadcast” across the larger array so that they have compatible shapes. Broadcasting provides a means of vectorizing array operations so that looping occurs in C instead of Python. It does this without making needless copies of data and usually leads to efficient algorithm implementations. There are, however, cases where broadcasting is a bad idea because it leads to inefficient use of memory that slows computation.

NumPy operations are usually done on pairs of arrays on an element-by-element basis. In the simplest case, the two arrays must have exactly the same shape, as in the following example:

```python
a = np.array([1.0, 2.0, 3.0])
b = np.array([2.0, 2.0, 2.0])
a * b
```

NumPy’s broadcasting rule relaxes this constraint when the arrays’ shapes meet certain constraints. The simplest broadcasting example occurs when an array and a scalar value are combined in an operation:

```python
a = np.array([1.0, 2.0, 3.0])
b = 2.0
a * b
```

The result is equivalent to the previous example where b was an array. We can think of the scalar b being stretched during the arithmetic operation into an array with the same shape as a. The new elements in b are simply copies of the original scalar. The stretching analogy is only conceptual. NumPy is smart enough to use the original scalar value without actually making copies, so that broadcasting operations are as memory and computationally efficient as possible.

The code in the second example is more efficient than that in the first because broadcasting moves less memory around during the multiplication (b is a scalar rather than an array).

#### General Broadcasting Rules

When operating on two arrays, NumPy compares their shapes element-wise. It starts with the trailing dimensions, and works its way forward. Two dimensions are compatible when

they are equal, or
one of them is 1
If these conditions are not met, a ValueError: frames are not aligned exception is thrown, indicating that the arrays have incompatible shapes. The size of the resulting array is the maximum size along each dimension of the input arrays.

Arrays do not need to have the same number of dimensions. For example, if you have a 256x256x3 array of RGB values, and you want to scale each color in the image by a different value, you can multiply the image by a one-dimensional array with 3 values. ### Manipulating indices

Proper index manipulation will gradually unleash the Python Jedi in you. Let's see how.

```python
## We will generate an 100x100 image with a star in it, having a Gaussian PSF
## Making the indices for easy manipulation
x,y = np.indices((100,100))
img_nonoise = 10*np.exp(-0.5*((x-40.)**2/4.**2 + (y-61.)**2/6.**2))
## Let's add some noise and a random offset
img = img_nonoise + np.random.normal(scale=1., size=(100,100)) + 4.321
```

Now imagine we want to sum all the pixels falling within a radio of 5 pixels from the center.

```python
## We calculate the distance from the center
dist = np.sqrt((x-40.)**2 + (y-61.)**2)
## We find the list that are within 5 pixels
inds_aper = dist <= 5.

## Performing the sum on those pixels
flux = (img[inds_aper]).sum()
print("flux (raw): {:.3f}".format(flux))

## Because of the offset, there is a sky/bias we need to remove. Let's calculated using the median value outside a circle of radius 10 pixels
inds_sky = dist <= 10
sky = np.median(img[~inds_sky])
print("sky: {:.3f}".format(sky))
## Note that above, dist > 10 would have work, but I wanted to show the use of "~" for the not condition

## The actual sky contribution to the flux is npixels*sky
flux1 = flux - inds_aper.sum()*sky
print("flux (sky subtracted): {:.3f}".format(flux1))

## Typically, the sky is calculated using a ring surrounding the aperture
inds_sky = (dist >= 7) * (dist <= 10)
sky = np.median(img[inds_sky])
print("sky (ring): {:.3f}".format(sky))
flux2 = flux - inds_aper.sum()*sky
print("flux (sky ring subtracted): {:.3f}".format(flux2))
```

Let's imagine there are bad pixels on the image and we'd like to get rid of them.

```python
## Let's generate 50 fake bad pixels, which would just be "hot pixels" here.
## First we need their coordinates, which we decide randomly
x_bad, y_bad = np.random.random_integers(0, 99, size=(2,50))
## Their values will be drawn from a log-normal distribution
val_bad = np.random.lognormal(mean=4, sigma=0.3, size=50)
img_bad = img.copy()
img_bad[x_bad,y_bad] = val_bad
```

Note that identifying the bad pixels might require some thinking. For optical astronomy, bad pixels would always fall at the same location of a detector. If one makes several 0-s exposures with the shutter closed, the image will have random thermal noise (low counts) and the bad pixels (high counts). One can stack them into a cube, calculate the median in order to uniformise the image, and then reject the bad pixels base on a threshold criterium.

```python
## Generating 20 low noise images
darks = np.random.normal(loc=1., scale=1., size=(10,100,100))
## On these images, we add the random hot pixels
darks[:,x_bad,y_bad] += np.random.lognormal(mean=4, sigma=0.3, size=(10,50))
```

```python
from IPython.display import clear_output
import time
fig, ax = pylab.subplots(figsize=(6,6))
for dark in darks:
    ax.cla()
    ax.imshow(dark, vmin=0, vmax=30, interpolation='nearest')
    display(fig)
    time.sleep(0.2)
    clear_output()
```

We will calculate the median value, along the first dimension of the image cube which has dimensions (nimg, nx, ny). That is, for each pixel we calculate the median along the image number axis.

```python
darks_median = np.median(darks, axis=0)
print("dimension of the median image of the darks: ", darks_median.shape)
fig, ax = pylab.subplots(figsize=(8,8))
ax.imshow(darks_median, interpolation='nearest')
inds_bad = darks_median > 20
print("number of bad pixels: {}".format(inds_bad.sum()))
```

### Manipulating indices for Jedi

Now let's consider an image contain multiple stars... Can we actually perform aperture photometry without the need of for loop? YES! (if you're a Python Jedi) Let's see how.

#### Generating a fake image with 20 stars

```python
## We will generate an 1000x1000 image with a 20 stars in it, having a Gaussian PSF and various luminosities
## Making the indices for easy manipulation
nstars = 20
x,y = np.indices((1000,1000))

## Generating an image full of noise
img_noise = np.random.normal(scale=1., size=(1000,1000)) + 4.321

## Generating the list of star positions and luminosities.
xpos = np.random.randint(30, 969, size=nstars)
ypos = np.random.randint(30, 969, size=nstars)
mag = np.random.lognormal(mean=2.5, sigma=0.4, size=nstars)

## Create a convenience function to make a star
sigma = 3.
make_star = lambda x0, y0, amp: amp * np.exp(-0.5/sigma**2*((x-x0)**2 + (y-y0)**2))

## Making the star by looping through the list and adding them to the noise image
img = img_noise.copy()
for i in xrange(nstars):
    #print("Making star {} at ({:.2f},{:.2f}) with peak flux {:.2f}".format(i, xpos[i], ypos[i], mag[i]))
    img += make_star(xpos[i], ypos[i], mag[i])

## Showing the image...
fig, ax = pylab.subplots(figsize=(8,8))
ax.imshow(img, interpolation='nearest')
```

#### Aperture photometry using for loop

Our first guess would be to use for loops...

```python
t0 = time.time()

## We create an empty list to store the fluxes
raw_flux = []
sky = []
flux = []

## We loop through the star list
for i in xrange(nstars):
    ## Storing the position in variables
    x0 = xpos[i]
    y0 = ypos[i]

    ## We calculate the distance from the center
    dist = np.sqrt((x-x0)**2 + (y-y0)**2)
    ## We find the list that are within 5 pixels
    inds_aper = dist <= 5.
    
    ## Performing the sum on those pixels
    raw_flux.append( (img[inds_aper]).sum() )
    
    ## The sky is calculated using a ring surrounding the aperture
    inds_sky = (dist >= 7) * (dist <= 10)
    sky.append( np.mean(img[inds_sky]) )

    ## The flux is the raw flux minus the sky
    ## Note that inds_aper.sum() is constant for each loop, so it could be precalculated
    flux.append( raw_flux[-1] - inds_aper.sum()*sky[-1] )

    ## Printing the result
    print("Star {} at ({:.2f},{:.2f}) has integrated flux {:.2f}".format(i+1, x0, y0, flux[-1]))

t1 = time.time()
print("Total run time is {:.3}".format(t1-t0))
```

Now let's redo the same thing but using Jedi techniques

```python
t0 = time.time()

## We calculate the distance from the center for all stars at once
print("img has dimensions:", img.shape)
print("x and y have dimensions:", x.shape)
print("xpos has dimensions:", xpos.shape)
print("xpos can be tweaked by adding trailing dimensions to broadcast:", xpos[:,np.newaxis,np.newaxis].shape)
dist = np.sqrt((x-xpos[:,np.newaxis,np.newaxis])**2 + (y-ypos[:,np.newaxis,np.newaxis])**2)
print("dist has dimensions:", dist.shape)

## We find the list that are within 5 pixels; this doesn't change
inds_aper = dist <= 5.
print("inds_aper has dimensions:", dist.shape)

## Performing the sum on those pixels that are within the aperture
## We only sum over the two last axis because the first contains the stars
#raw_flux_new = (img*inds_aper).sum(axis=-1).sum(axis=-1)
raw_flux_new = (img*inds_aper).T.sum(axis=0).sum(axis=0)
print("raw_flux_new has dimensions:", raw_flux_new.shape)

## The sky is calculated using a ring surrounding the aperture
## Here we have to switch method for the sky calculation because the median is not a separable process
inds_sky = (dist >= 7) * (dist <= 10)
sky_new = (inds_sky*img).sum(axis=-1).sum(axis=-1) / (inds_sky).sum(axis=-1).sum(axis=-1)
print("sky_new has dimensions:", sky_new.shape)

## The flux is the raw flux minus the sky
## Note that inds_aper.sum() is constant for each loop, so it could be precalculated
flux_new = raw_flux_new - sky_new*inds_aper.sum(axis=-1).sum(axis=-1)

t1 = time.time()
print("Total run time is {:.3}".format(t1-t0))

## Printing the result
for i in xrange(nstars):
    print("Star {} at ({:.2f},{:.2f}) has integrated flux {:.2f}".format(i+1, xpos[i], ypos[i], flux_new[i]))
```
